```python
# import libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# show the image
plt.imshow(img)
```

```python
# Matplotlib  --> RGB Channel order
# OpenCV --> BGR Channel order 
```

```python
# fix the image to RGB order
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
## read image as grayscale
dog_gray = cv2.imread('Data/puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
# shape of the image
dog_gray.shape
```

```python
# resizing the image
res_img = cv2.resize(fix_img,(1000,1000))
plt.imshow(res_img)
```

```python
## flip along different axis
plt.imshow(cv2.flip(fix_img,-1))
```

```python
# create a blank image
blank = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank)
```

```python
#draw rectangle on the blank image
#pt1 - top left corner
#pt2 - bottom right corner
cv2.rectangle(blank,pt1 =(200,200),pt2=(400,400),color=(255,225,255),thickness=5)
```

```python
plt.imshow(blank)
```

```python
## draw circles
cv2.circle(blank,center=(256,256),radius=30,color=(0,0,255),thickness=5)
```

```python
plt.imshow(blank)
```

```python
## to draw filled in circles are images give thickness as negative 1(-1)

cv2.circle(blank,center=(256,256),radius=30,color=(0,0,255),thickness=-1)
```

```python
plt.imshow(blank)
```

```python
## to draw line
cv2.line(blank,pt1=(0,0),pt2=(512,512),color=(255,0,0),thickness=3)
```

```python
plt.imshow(blank)
```

```python
## add text to image
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(blank,text="HEY..!!",org=(10,400),fontFace=font,fontScale=1,color=(255,255,255),thickness=2)
```

```python
plt.imshow(blank)
```

```python
new_blank = np.zeros(shape=(512,512,3),dtype=np.int32)
```
